Join Station ID to Zip Code CSV, and only filter the columns that are actually needed.

```python
%%sql

us_data_all_df <<

SELECT z.station_id,
       z.zip_code,
       a.*
FROM read_csv('artifacts/us_data_all_zip.csv') z
    LEFT JOIN read_csv('artifacts/us_data.csv') a
        ON z.station_id = a.STATION
LIMIT 10
```
